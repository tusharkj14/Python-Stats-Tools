```python
import numpy as np
import pandas as pd
```

```python
def covar(data1, data2):
    x_mean = np.mean(data1)
    y_mean = np.mean(data2)
    i, s = 0, []
    for i in range(len(data1)):
        p = (data1[i]-x_mean)*(data2[i]-y_mean)
        s.append(p)
    return (sum(s)/len(data1))
```

```python
def pcoeff(data):
    x = data[0]
    y = data[1]
    std_x = np.std(x)
    std_y = np.std(y)
    pcoeff_ =  covar(x, y)/(std_x*std_y)
    return np.round(pcoeff_, decimals=5)
```

```python
def send_data(x, y, a):
    if(a == '1'):
        d1 = [float(i) for i in x]
        d2 = [float(i) for i in y]
        print(covar(d1, d2))
    elif(a == '2'):
        d1 = [float(i) for i in x]
        d2 = [float(i) for i in y]
        d = [d1, d2]
        print(pcoeff(d))
```
